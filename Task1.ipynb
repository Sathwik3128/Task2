```python
marks = [45, 78, 90, 33, 60]
pass_count = sum(1 for mark in marks if mark >= 50)
fail_count = len(marks) - pass_count
print(f"Total Pass Students: {pass_count}")
print(f"Total Fail Students: {fail_count}")
print("Final Result:", "Pass" if pass_count > fail_count else "Fail")
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]

print(cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"'{msg}' length: {length}")
    if length > 10:
        print("  -> FLAGGED: Message too long!")
```

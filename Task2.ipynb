```python
#Unique words in a sentence
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Highest Salary Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_name = ""
highest_salary = 0
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name
print("Highest Salary:", highest_name, "-", highest_salary)
```

```python
#Find Minimum and Maximum Values
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count = count + 1
print("Products above 1000:", count)
```

```python
#Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P","A"]
present_count = 0
total_days = len(attendance)
for day in attendance:
    if day == "P":
        present_count = present_count + 1
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

```python
#Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
#Count Character Frequency
text = "sathwik"
frequency = {}
for char in text:
    if char in frequency:
        frequency[char] = frequency[char] + 1
    else:
        frequency[char] = 1
print(frequency)
```

```python
#Check if a Key Exists in a Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
key = "Ravi"
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
